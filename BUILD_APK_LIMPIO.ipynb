## 2Ô∏è‚É£ Subir proyecto (ZIP)

```python
from google.colab import files
import zipfile
import os

# Limpiar todo primero
print("üßπ Limpiando archivos anteriores...")
!rm -rf /content/app
!rm -rf ~/.buildozer

# Crear directorio limpio
!mkdir -p /content/app
%cd /content/app

print("\nüì¶ Sube el ZIP del proyecto:")
uploaded = files.upload()

zip_filename = list(uploaded.keys())[0]
print(f"\nüìÇ Descomprimiendo {zip_filename}...")

with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
    zip_ref.extractall('.')

os.remove(zip_filename)

print("\n‚úÖ Proyecto descomprimido")
print("\nüìÇ Archivos:")
!ls -lh

if os.path.exists('main.py') and os.path.exists('buildozer.spec'):
    print("\n‚úÖ Archivos principales encontrados")
    # Mostrar versi√≥n
    !grep "version =" buildozer.spec
else:
    print("\n‚ùå ERROR: Faltan archivos principales")
```

## 5Ô∏è‚É£ Descargar APK

```python
from google.colab import files
import os
import glob

# Buscar el APK
apk_files = glob.glob('bin/*.apk')

if apk_files:
    apk_path = apk_files[0]
    print(f"üì¶ Descargando: {apk_path}")
    print(f"üìä Tama√±o: {os.path.getsize(apk_path) / (1024*1024):.2f} MB\n")
    
    files.download(apk_path)
    
    print("\n‚úÖ APK descargado")
    print("\nüì± Para instalar:")
    print("   1. Transfiere el APK a tu Android")
    print("   2. Activa 'Instalar apps desconocidas'")
    print("   3. Abre el APK y presiona 'Instalar'")
else:
    print("‚ùå No se encontr√≥ el APK")
    print("Verifica que la compilaci√≥n se complet√≥ correctamente")
```
